```python
import random
import numpy as np
import timeit
```

```python
randomindices = [random.randint(0, 10000) for i in range(100000)]
randomlenindi = [random.randint(0, 100) for i in range(100000)]
randomlengths = [random.randint(0, 10000) for i in range(100000)]
zipper = list(zip(randomlengths, randomindices, randomlenindi))
```

```python
print("Numpy argsort")
print(timeit.timeit('np.argsort(randomlengths)', globals=globals(), number=1000))
print("Python .sort key")
print(timeit.timeit('zipper.sort(key=lambda x: x[1])', globals=globals(), number=1000))
print("Python sorted key")
print(timeit.timeit('sorted(zipper, key=lambda x: x[1])', globals=globals(), number=1000))
print("Python .sort w/o key")
print(timeit.timeit('zipper.sort()', globals=globals(), number=1000))
```

```python
print("Python .sort")
print(timeit.timeit('zipper.sort()', globals=globals(), number=1000))
print("Python .sort")
print(timeit.timeit('zipper.sort(key=lambda x: x[0])', globals=globals(), number=1000))
print("Python .sort")
print(timeit.timeit('randomlengths.sort()', globals=globals(), number=1000))
```
